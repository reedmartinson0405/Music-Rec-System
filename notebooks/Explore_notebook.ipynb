Getting rid of the brackets and quotes around the artist name and artist ID. 

```python
df_tracks['artist_id'] = df_tracks['id_artists'].str.strip("[]'").str.split(',')
```

```python
df_track2 = df_tracks.explode('artist_id')
```

```python
df_track2.head()
```

```python
save_matt = pd.merge(df_track2, df_artists, right_on='id', left_on= 'artist_id')
```

```python
save_matt.columns
```

```python
save_matt.head(1)
```

Getting rid of repetitive columns. 

```python
save_matt.drop(columns= ['id_y', 'id_artists', 'artists'], inplace=True)
```

```python
save_matt.head(1)
```

```python
save_matt.rename(columns={'id_x':'track_id', 'name_x':'track_name',
                          'popularity_x':'track_popularity', 
                          'name_y':'artist_name', 'popularity_y':'artist_popularity'}, inplace = True)
```

```python
save_matt
```

```python
save_matt.columns[save_matt.applymap(type).eq(list).any()]
```

```python
save_matt.applymap(lambda x: isinstance(x,list)).any()
```

Getting rid of more brackets and apostrophes that are still wrapped around the genre text.

```python
save_matt['genres'] = save_matt['genres'].str.strip("[]'")
```

```python
save_matt['genres']
```

```python
save_matt['genres'] = save_matt['genres'].str.replace("'", '')
```
